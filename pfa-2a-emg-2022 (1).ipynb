```python

#déphasage

def dephasage(signa1,signal2):
     
        
        h1= hilbert(signal1)
        h2= hilbert (signal2)

        P1=np.unwrap(np.angle(h1))
        P2=np.unwrap(np.angle(h2))

        dephasage = np.mean(P2-P1)
        return dephasage
    
    
#test

signal1=df.loc[:,'elec1']    
signal2=df.loc[:,'elec2'] 
dephasage(signal1,signal2)
```

```python

#frequence

def frequence(values):
    
        spectrum = fft.fft(values)
        freq = fft.fftfreq(len(spectrum))
        plt.plot(freq, abs(spectrum))
        threshold = 0.5 * max(abs(spectrum))
        mask = abs(spectrum) > threshold
        peaks = freq[mask]
        return peaks
    
values=df.loc[:,'elec1']     ()
frequence(values)
```

```python
def descriptor(file):
    df=echantillonage(file)
    df_desc = pd.DataFrame()

    for c in df.columns:
        df_desc[c] = signal.cwt(df[c].values, signal.ricker, np.arange(1, 6)).reshape(-1, 1)[:, 0]
    
    return df_desc
```
